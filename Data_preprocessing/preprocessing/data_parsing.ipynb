```python
review=[]
for line in data_file:
    #line = data_file.readline()
    review.append(json.loads(line))
review_df = pd.DataFrame(review)
data_file.close()
```

```python
review_df.head()
```

```python
review_df.to_csv("review_df.csv",index=False)
```

```python
review_df["sum_ufc"]=review_df["useful"]+review_df["funny"]+review_df["cool"]
```

```python
review_df_withufc=review_df[review_df["sum_ufc"]>0]
```

```python
review_df_withufc.shape
```

```python
review_df_non=review_df[review_df["useful"].apply(lambda x: x==0) & 
                        review_df["funny"].apply(lambda x: x==0) & 
                        review_df["cool"].apply(lambda x: x==0)]
```

```python
review_df_non_s=review_df_non
```

```python
review_df_non_s.loc[review_df_non_s["funny"] <= 10, "funny_label"] =0
```

```python
review_df_non_s.loc[review_df_non_s["useful"] <= 10, "useful_label"] =0
```

```python
review_df_non_s.loc[review_df_non_s["cool"] <= 10, "cool_label"] =0
```

```python
review_df_non_s_len=review_df_non_s[review_df_non_s["text"].apply(lambda x: len(x)>256)]
```

```python
review_df_non_s_len_s=review_df_non_s_len.sample(n=10000)
```

```python
review_df_non_s_len_s.shape
```

```python
review_df_u=review_df[review_df["useful"]>10]
```

```python
review_df_u.shape
```

```python
review_df_f=review_df[review_df["funny"]>10]
```

```python
review_df_f.shape
```

```python
review_df_c=review_df[review_df["cool"]>10]
```

```python
review_df_c.shape
```

```python
review_df_u
```

```python
review_ufc=pd.concat([review_df_u,review_df_f,review_df_c])
```

```python
review_ufc.loc[review_ufc["useful"] > 10, "useful_label"] =1
review_ufc.loc[review_ufc["useful"] <= 10, "useful_label"] =0
```

```python
review_ufc.loc[review_ufc["funny"] > 10, "funny_label"] =1
review_ufc.loc[review_ufc["funny"] <= 10, "funny_label"] =0
```

```python
review_ufc.loc[review_ufc["cool"] > 10, "cool_label"] =1
review_ufc.loc[review_ufc["cool"] <= 10, "cool_label"] =0
```

```python
review_ufc
```

```python
review_ufc_unique=review_ufc.drop_duplicates()
```

```python
review_ufc_unique.shape
```

```python
review_ufc_unique.reset_index(inplace=True,drop=True)
```

```python
review_ufc_unique.shape
```

```python
review_ufc_len=review_ufc_unique[review_ufc_unique["text"].apply(lambda x: len(x)>256)]
```

```python
review_ufc_len.shape
```

```python
review_ufc_len.head(5)
```

```python
review_ufc_len_u=review_ufc_len[review_ufc_len["useful_label"].apply(lambda x: x==1) & 
                                review_ufc_len["funny_label"].apply(lambda x: x==0)& 
                                review_ufc_len["cool_label"].apply(lambda x: x==0)]
```

```python
review_ufc_len_u.shape
```

```python
review_ufc_len_uf=review_ufc_len[review_ufc_len["useful_label"].apply(lambda x: x==1) & 
                                review_ufc_len["funny_label"].apply(lambda x: x==1)& 
                                review_ufc_len["cool_label"].apply(lambda x: x==0)]
```

```python
review_ufc_len_uf.shape
```

```python
review_ufc_len_ufc=review_ufc_len[review_ufc_len["useful_label"].apply(lambda x: x==1) & 
                                review_ufc_len["funny_label"].apply(lambda x: x==1)& 
                                review_ufc_len["cool_label"].apply(lambda x: x==1)]
```

```python
review_ufc_len_ufc.shape
```

```python
review_ufc_len_uc=review_ufc_len[review_ufc_len["useful_label"].apply(lambda x: x==1) & 
                                review_ufc_len["funny_label"].apply(lambda x: x==0)& 
                                review_ufc_len["cool_label"].apply(lambda x: x==1)]
```

```python
review_ufc_len_uc.head(5)
```

```python
review_ufc_len_uc.shape
```

```python
review_ufc_len_fc=review_ufc_len[review_ufc_len["useful_label"].apply(lambda x: x==0) & 
                                review_ufc_len["funny_label"].apply(lambda x: x==1)& 
                                review_ufc_len["cool_label"].apply(lambda x: x==1)]
```

```python
review_ufc_len_fc.shape
```

```python
review_ufc_len_f=review_ufc_len[review_ufc_len["useful_label"].apply(lambda x: x==0) & 
                                review_ufc_len["funny_label"].apply(lambda x: x==1)& 
                                review_ufc_len["cool_label"].apply(lambda x: x==0)]
```

```python
review_ufc_len_f.shape
```

```python
review_ufc_len_c=review_ufc_len[review_ufc_len["useful_label"].apply(lambda x: x==0) & 
                                review_ufc_len["funny_label"].apply(lambda x: x==0)& 
                                review_ufc_len["cool_label"].apply(lambda x: x==1)]
```

```python
review_ufc_len_c.shape
```

```python
review_ufc_len_u_sel=review_ufc_len_u.sample(n=20000)
```

```python
review_cleaned=pd.concat([review_df_non_s_len_s,review_ufc_len_u_sel,review_ufc_len_uf,
                          review_ufc_len_uc,review_ufc_len_ufc,review_ufc_len_f,
                         review_ufc_len_fc,review_ufc_len_c],sort=False)
```
